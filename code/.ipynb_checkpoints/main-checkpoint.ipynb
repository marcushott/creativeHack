```python
from data_utils import compute_norm_params

mean, dev, dmax, dmin = compute_norm_params(data_file='../datasets/processed/train.json')

print("Mean and standard deviation over all channels:")
print(mean)
print(dev)
print("Min and max values in the training set:")
print(dmin)
print(dmax)
```

Load Data

```python
from data_utils import IcebergData
from torchvision import transforms

data_transform = transforms.Compose([
         transforms.ToPILImage(),
         #transforms.RandomHorizontalFlip(),
         #transforms.RandomVerticalFlip(),
         #transforms.RandomRotation(degrees=(-30,30)),
         #transforms.RandomCrop(65),
         #transforms.Resize((75, 75)),
         transforms.ToTensor(),
])

norm_transform = transforms.Normalize(mean=mean, std=dev)

train_data = IcebergData(data_file='../datasets/processed/train.json', start=0, end=1500, 
                         transform = data_transform, norm_transform = norm_transform)
validate_data = IcebergData(data_file='../datasets/processed/train.json',start=1500, end=1604)
print("Number of training samples: ", train_data.__len__())
```

Train Network

```python
import torch
from solver import Solver
from classification_nn import ClassificationNN


train_loader = torch.utils.data.DataLoader(train_data, batch_size=25, shuffle=True, num_workers=4)
val_loader = torch.utils.data.DataLoader(validate_data, batch_size=25, shuffle=True, num_workers=4)

model = ClassificationNN()
solver = Solver(optim_args={"lr": 1e-3, "weight_decay": 0.000})
solver.train(model, train_loader, val_loader, log_nth=0, num_epochs=30)


```

```python
import matplotlib.pyplot as plt
plt.subplot(2, 1, 1)

plt.plot(solver.train_loss_history, 'o')
plt.plot(solver.val_loss_history, 'o')
plt.xlabel('iteration')
plt.ylabel('loss')
plt.legend(['train', 'val'], loc='upper left')

plt.subplot(2, 1, 2)
plt.plot(solver.train_acc_history, '-o')
plt.plot(solver.val_acc_history, '-o')
plt.legend(['train', 'val'], loc='upper left')
plt.xlabel('epoch')
plt.ylabel('accuracy')

plt.show()

```

Test Model

```python
from torch.autograd import Variable
import torch

test_data = IcebergData(data_file='../datasets/processed/test.json')
test_loader = torch.utils.data.DataLoader(test_data,
                                          batch_size=1,
                                          shuffle=False,
                                          num_workers=1)

test_results = []
model.eval()

for inputs, id in test_loader:
    inputs = Variable(inputs)
    
    if model.is_cuda:
        inputs = inputs.cuda()
    
    prob = torch.nn.Softmax(dim = 1)
    outputs = prob(model.forward(inputs))
    outputs = outputs.data.numpy()
    
    test_results.append((id[0], outputs[0,1]))
    print(test_results[-1])

    

model.train()
```

Save results

```python
import csv

csvfile = "../datasets/output.csv"
with open(csvfile, "w") as output:
    writer = csv.writer(output)
    writer.writerows(test_results)
```
